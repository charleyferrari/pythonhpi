```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline

import plotly.plotly as py
import plotly.graph_objs as go

import cufflinks as cf
```

```python
hp = pd.read_csv('hpipy.csv')
```

```python
hp['DATE'] = pd.to_datetime(hp['DATE'])

city = 'Boston'

hpcity = hp[(hp['Metro'] == city) & (hp['Seasonality'] == 'SA')]

hpcity = hpcity.pivot(index = 'DATE', columns = 'Tier', values = 'HPI')

hpcity = hpcity[['High', 'Middle', 'Low']]
```

```python
hpplot = hpcity.plot(title = city + " HPI")
```

```python
layout = dict(title = city + ' HPI',
              yaxis = dict(title = 'HPI'),
              xaxis = dict(title = 'Date')
             )

hpcity.iplot(layout = layout, filename = 'hpi/HPI Python Overwrite')
```
